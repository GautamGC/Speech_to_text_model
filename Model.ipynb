```python
import torch
import torchaudio
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import Dataset, DataLoader
from torchaudio.datasets import LIBRISPEECH
from torchaudio.transforms import MelSpectrogram, FrequencyMasking, TimeMasking
from tqdm.notebook import tqdm
import os
import time
import soundfile as sf
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# Configuration class
class Config:
    # Data parameters
    base_dir = "./data"
    train_url = "train-clean-100"  # LibriSpeech subset
    test_url = "test-clean"
    
    # Audio processing parameters
    sample_rate = 16000
    n_mels = 128  # Number of mel filterbanks
    n_fft = 2048  # FFT size
    win_length = 400  # Window size
    hop_length = 160  # Hop size
    
    # Model parameters
    hidden_size = 256
    num_layers = 3
    dropout = 0.1
    vocab_size = 29  # 26 letters + space + blank + apostrophe
    
    # Training parameters
    batch_size = 16
    learning_rate = 3e-4
    epochs = 20
    early_stopping_patience = 5

# Create configuration
config = Config()
# Create data directory if it doesn't exist
os.makedirs(config.base_dir, exist_ok=True)
```

```python
def create_vocabulary():
    """Create character mapping dictionaries"""
    char_map = {}
    index_map = {}
    
    # Add blank symbol (used by CTC)
    index_map[0] = '<blank>'
    char_map['<blank>'] = 0
    
    # Add lowercase letters, space, and apostrophe
    for index, char in enumerate("abcdefghijklmnopqrstuvwxyz '", 1):
        char_map[char] = index
        index_map[index] = char
    
    return char_map, index_map
```

```python
class AudioPreprocessor:
    """Class to preprocess audio into mel spectrograms"""
    def __init__(self, config):
        self.mel_transform = MelSpectrogram(
            sample_rate=config.sample_rate,
            n_fft=config.n_fft,
            win_length=config.win_length,
            hop_length=config.hop_length,
            n_mels=config.n_mels
        )
        self.freq_mask = FrequencyMasking(freq_mask_param=15)
        self.time_mask = TimeMasking(time_mask_param=35)
        self.training = True
    
    def __call__(self, audio_tensor):
        # Convert to mono if stereo
        if audio_tensor.shape[0] > 1:
            audio_tensor = torch.mean(audio_tensor, dim=0, keepdim=True)
        
        # Convert to mel spectrogram
        mel_spec = self.mel_transform(audio_tensor)
        
        # Convert to log mel spectrogram
        log_mel = torch.log1p(mel_spec)
        
        # Apply data augmentation (only during training)
        if self.training:
            log_mel = self.freq_mask(log_mel)
            log_mel = self.time_mask(log_mel)
        
        # Normalize
        mean = log_mel.mean()
        std = log_mel.std()
        log_mel = (log_mel - mean) / (std + 1e-8)
        
        return log_mel
```

```python
class SpeechDataset(Dataset):
    """Dataset for LibriSpeech data"""
    def __init__(self, librispeech_path, subset, config, char_map, preprocessor, train=True):
        self.dataset = LIBRISPEECH(librispeech_path, url=subset, download=True)
        self.char_map = char_map
        self.preprocessor = preprocessor
        self.train = train
        self.preprocessor.training = train
    
    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        waveform, sample_rate, transcript, _, _, _ = self.dataset[idx]
        
        # Resample if needed
        if sample_rate != 16000:
            waveform = torchaudio.transforms.Resample(sample_rate, 16000)(waveform)
        
        # Preprocess audio
        feature = self.preprocessor(waveform)
        
        # Convert transcript to lowercase and map characters to indices
        transcript = transcript.lower()
        target = torch.tensor([self.char_map.get(c, self.char_map[' ']) for c in transcript if c in self.char_map])
        
        return feature, target, feature.shape[2]  # feature, target, input_length
```

```python
def collate_fn(batch):
    """Custom collate function for batching variable length sequences"""
    features = []
    targets = []
    input_lengths = []
    target_lengths = []
    
    # Sort batch by feature length (descending)
    batch.sort(key=lambda x: x[2], reverse=True)
    
    for feature, target, input_length in batch:
        features.append(feature)
        targets.append(target)
        input_lengths.append(input_length)
        target_lengths.append(len(target))
    
    # Pad features
    max_input_length = max(input_lengths)
    padded_features = []
    for feature in features:
        padded_feature = torch.zeros(1, feature.shape[1], max_input_length)
        padded_feature[0, :, :feature.shape[2]] = feature
        padded_features.append(padded_feature)
    
    # Stack and create tensors
    features = torch.cat(padded_features, dim=0)
    targets = torch.cat(targets)
    input_lengths = torch.tensor(input_lengths)
    target_lengths = torch.tensor(target_lengths)
    
    return features, targets, input_lengths, target_lengths
```

```python
class SpeechModel(nn.Module):
    """RNN-based speech recognition model"""
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim, dropout):
        super(SpeechModel, self).__init__()
        
        # Bidirectional RNN layers
        self.rnn = nn.GRU(
            input_size=input_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0,
            batch_first=True
        )
        
        # Fully connected layer
        self.fc = nn.Linear(hidden_dim * 2, output_dim)  # *2 because bidirectional
    
    def forward(self, x, input_lengths):
        # x shape: (batch, 1, n_mels, time)
        # Reshape to (batch, time, n_mels)
        batch_size = x.size(0)
        x = x.squeeze(1).transpose(1, 2)
        
        # Pack padded sequence
        packed_x = nn.utils.rnn.pack_padded_sequence(
            x, input_lengths.cpu(), batch_first=True, enforce_sorted=True
        )
        
        # Apply RNN
        packed_output, _ = self.rnn(packed_x)
        
        # Unpack
        output, _ = nn.utils.rnn.pad_packed_sequence(packed_output, batch_first=True)
        
        # Apply final layer
        output = self.fc(output)
        
        # Apply log softmax
        output = F.log_softmax(output, dim=2)
        
        return output
```

```python
class ASRModel(nn.Module):
    """RNN-based speech recognition model that matches the saved checkpoint architecture"""
    def __init__(self):
        super(ASRModel, self).__init__()
        
        # Bidirectional RNN layers
        self.rnn = nn.GRU(
            input_size=128,        # n_mels from the original config
            hidden_size=256,       # hidden_size from the original config
            num_layers=3,          # num_layers from the original config
            bidirectional=True,
            dropout=0.1,           # dropout from the original config
            batch_first=True
        )
        
        # Fully connected layer - needs to be 512 input (256*2 for bidirectional)
        self.fc = nn.Linear(256 * 2, 29)  # 29 is vocab_size from original config
    
    def forward(self, x, input_lengths):
        # x shape: (batch, 1, n_mels, time)
        # Reshape to (batch, time, n_mels)
        batch_size = x.size(0)
        x = x.squeeze(1).transpose(1, 2)
        
        # Pack padded sequence
        packed_x = nn.utils.rnn.pack_padded_sequence(
            x, input_lengths.cpu(), batch_first=True, enforce_sorted=True
        )
        
        # Apply RNN
        packed_output, _ = self.rnn(packed_x)
        
        # Unpack
        output, _ = nn.utils.rnn.pad_packed_sequence(packed_output, batch_first=True)
        
        # Apply final layer
        output = self.fc(output)
        
        # Apply log softmax
        output = F.log_softmax(output, dim=2)
        
        return output
```

```python
def decode_prediction(output, index_map):
    """Decode model prediction to text"""
    # Output shape: (time, vocab_size)
    # Get most likely class at each timestamp
    predictions = torch.argmax(output, dim=1)
    
    # Convert to list and remove duplicates and blanks
    decoded = []
    prev = -1
    for p in predictions:
        if p != 0 and p != prev:  # 0 is blank
            decoded.append(index_map[p.item()])
        prev = p
    
    # Join characters to form string
    return ''.join(decoded)
```

```python
def levenshtein_distance(s, t):
    """Calculate Levenshtein distance between two word lists"""
    m, n = len(s), len(t)
    d = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(m + 1):
        d[i][0] = i
    for j in range(n + 1):
        d[0][j] = j
    
    for j in range(1, n + 1):
        for i in range(1, m + 1):
            if s[i - 1] == t[j - 1]:
                d[i][j] = d[i - 1][j - 1]
            else:
                d[i][j] = min(d[i - 1][j], d[i][j - 1], d[i - 1][j - 1]) + 1
    
    return d[m][n]
```

```python
def train_epoch(model, train_loader, criterion, optimizer, device, epoch, log_freq=5):
    """Train model for one epoch with more frequent updates"""
    print("Starting training epoch...")
    epoch_start_time = time.time()  # Initialize the time at the beginning of the function
    model.train()
    total_loss = 0
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch} [Train]")
    
    for batch_idx, (features, targets, input_lengths, target_lengths) in enumerate(progress_bar):
        print(f"Processing batch {batch_idx}, features shape: {features.shape}")
        features = features.to(device)
        targets = targets.to(device)
        input_lengths = input_lengths.to(device)
        target_lengths = target_lengths.to(device)
        
        optimizer.zero_grad()
        
        # Forward pass
        outputs = model(features, input_lengths)
        outputs = outputs.transpose(0, 1)  # (time, batch, vocab_size)
        
        # Calculate loss
        loss = criterion(outputs, targets, input_lengths, target_lengths)
        
        # Backward and optimize
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5)  # Gradient clipping
        optimizer.step()
        
        total_loss += loss.item()
        
        # More frequent updates
        if batch_idx % log_freq == 0 or batch_idx == len(train_loader) - 1:
            current_lr = optimizer.param_groups[0]['lr']
            elapsed = time.time() - epoch_start_time
            progress_bar.set_postfix({
                "loss": f"{loss.item():.4f}", 
                "avg_loss": f"{total_loss/(batch_idx+1):.4f}",
                "lr": f"{current_lr:.6f}",
                "time": f"{elapsed:.1f}s"
            })
            
            # For Jupyter: force update
            progress_bar.refresh()
        
        # For Jupyter: update more frequently 
        if batch_idx % 2 == 0:
            progress_bar.update(min(2, len(train_loader) - batch_idx - 1))
    
    avg_loss = total_loss / len(train_loader)
    print(f'Epoch {epoch}, Average Train Loss: {avg_loss:.4f}')
    return avg_loss
```

```python
def evaluate(model, test_loader, criterion, device, index_map):
    """Evaluate model on test set"""
    model.eval()
    total_loss = 0
    total_wer = 0
    total_samples = 0
    progress_bar = tqdm(test_loader, desc="Evaluation")
    
    with torch.no_grad():
        for features, targets, input_lengths, target_lengths in progress_bar:
            features = features.to(device)
            targets = targets.to(device)
            input_lengths = input_lengths.to(device)
            target_lengths = target_lengths.to(device)
            
            # Forward pass
            outputs = model(features, input_lengths)
            outputs = outputs.transpose(0, 1)  # (time, batch, vocab_size)
            
            # Calculate loss
            loss = criterion(outputs, targets, input_lengths, target_lengths)
            total_loss += loss.item()
            
            # Decode predictions for WER calculation
            batch_size = features.size(0)
            for i in range(batch_size):
                output_length = input_lengths[i]
                output = outputs[:output_length, i, :]
                
                # Get target indices
                target_start = torch.sum(target_lengths[:i]).item()
                target_end = target_start + target_lengths[i].item()
                target_indices = targets[target_start:target_end]
                
                # Decode prediction
                predicted_text = decode_prediction(output, index_map)
                
                # Get target text
                target_text = ''.join([index_map[t.item()] for t in target_indices])
                
                # Calculate word error rate
                if len(target_text.split()) > 0:  # Avoid division by zero
                    words = target_text.split()
                    predicted_words = predicted_text.split()
                    wer = levenshtein_distance(words, predicted_words) / len(words)
                    total_wer += wer
                    total_samples += 1
            
    avg_loss = total_loss / len(test_loader)
    avg_wer = total_wer / max(1, total_samples)  # Avoid division by zero
    print(f'Test Loss: {avg_loss:.4f}, Word Error Rate: {avg_wer:.4f}')
    return avg_loss, avg_wer
```

```python
def transcribe_audio(model, audio_path, char_map, index_map, config, device):
    """Transcribe an audio file"""
    model.eval()
    
    # Initialize preprocessor
    preprocessor = AudioPreprocessor(config)
    preprocessor.training = False

    # Check if the file exists
    if not os.path.exists(audio_path):
        raise FileNotFoundError(f"Error: The file '{audio_path}' was not found.")

    # Check audio file format
    try:
        audio_info = sf.info(audio_path)
        print(f"Audio Info: {audio_info}")  # Debugging
    except Exception as e:
        raise RuntimeError(f"Error reading audio file: {e}")

    # Load audio
    try:
        waveform, sample_rate = torchaudio.load(audio_path)
    except Exception as e:
        raise RuntimeError(f"Error loading audio with torchaudio: {e}")

    # Display audio waveform
    plt.figure(figsize=(10, 4))
    plt.plot(waveform[0].numpy())
    plt.title('Audio Waveform')
    plt.show()
    
    # Resample if needed
    if sample_rate != config.sample_rate:
        waveform = torchaudio.transforms.Resample(sample_rate, config.sample_rate)(waveform)
    
    # Preprocess audio
    feature = preprocessor(waveform).to(device)
    input_length = torch.tensor([feature.shape[2]]).to(device)
    
    # Display mel spectrogram
    plt.figure(figsize=(10, 4))
    plt.imshow(feature[0].cpu().numpy(), aspect='auto', origin='lower')
    plt.title('Mel Spectrogram')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
    
    # Forward pass
    with torch.no_grad():
        output = model(feature.unsqueeze(0), input_length)
        output = output.transpose(0, 1)  # (time, batch, vocab_size)
        output = output[:, 0, :]  # Get first batch item
    
    # Decode prediction
    predicted_text = decode_prediction(output, index_map)
    return predicted_text
```

```python
try:
    # Create datasets
    train_dataset = SpeechDataset(config.base_dir, config.train_url, config, char_map, 
                                  preprocessor, train=True)
    
    # For demo purposes, we'll use a small sample
    sample_size = 7000  # Reduced from 100 for faster debugging
    indices = torch.randperm(len(train_dataset))[:sample_size]
    
    train_subset = torch.utils.data.Subset(train_dataset, indices)
    
    test_dataset = SpeechDataset(config.base_dir, config.test_url, config, char_map, 
                                preprocessor, train=False)
    
    test_subset = torch.utils.data.Subset(test_dataset, torch.randperm(len(test_dataset))[:100])
    
    # Create data loaders with reduced batch size
    config.batch_size = 10  # Reduced for faster processing
    
    train_loader = DataLoader(
        train_subset,
        batch_size=config.batch_size,
        shuffle=True,
        collate_fn=collate_fn,
        num_workers=0,  # Reduced for Jupyter notebook
        pin_memory=True
    )
    
    test_loader = DataLoader(
        test_subset,
        batch_size=config.batch_size,
        shuffle=False,
        collate_fn=collate_fn,
        num_workers=0,  # Reduced for Jupyter notebook
        pin_memory=True
    )
    
    print(f"Training on {len(train_subset)} samples")
    print(f"Testing on {len(test_subset)} samples")
    
    # Initialize model for training
    model = SpeechModel(
        input_dim=config.n_mels,
        hidden_dim=config.hidden_size,
        num_layers=config.num_layers,
        output_dim=config.vocab_size,
        dropout=config.dropout
    ).to(device)
    
    # Initialize loss function and optimizer
    criterion = nn.CTCLoss(blank=0, zero_infinity=True)
    optimizer = optim.Adam(model.parameters(), lr=config.learning_rate)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.5, patience=2
    )
    
    # Create directory for checkpoints
    os.makedirs("checkpoints", exist_ok=True)
    
    # Training loop
    train_losses = []
    test_losses = []
    wers = []
    best_wer = float('inf')
    patience_counter = 0
    
    # Reduce epochs for demonstration
    num_epochs = 15  # Reduced for faster completion
    
    # Add log frequency variables
    log_freq = 2  # Log more frequently
    
    # Overall progress bar for the entire training process
    overall_progress = tqdm(total=num_epochs, desc="Overall Training Progress", position=0)
    
    # Add time tracking
    start_time = time.time()
    epoch_start_time = start_time
    
    for epoch in range(1, num_epochs + 1):
        print(f"\n{'='*50}\nEpoch {epoch}/{num_epochs}\n{'='*50}")
        epoch_start_time = time.time()
        
        # Update description with current epoch
        overall_progress.set_description(f"Training Progress (Epoch {epoch}/{num_epochs})")
        
        # Train for one epoch with more frequent updates
        train_loss = train_epoch(model, train_loader, criterion, optimizer, device, epoch, log_freq)
        epoch_train_time = time.time() - epoch_start_time
        print(f"Train Loss: {train_loss:.4f} (Time: {epoch_train_time:.1f}s)")
        
        # Start evaluation timer
        eval_start_time = time.time()
        
        # Evaluate
        test_loss, wer = evaluate(model, test_loader, criterion, device, index_map)
        epoch_eval_time = time.time() - eval_start_time
        print(f"Test Loss: {test_loss:.4f}, WER: {wer:.4f} (Eval Time: {epoch_eval_time:.1f}s)")
        
        # Calculate and print time estimates
        epoch_total_time = time.time() - epoch_start_time
        elapsed_total = time.time() - start_time
        avg_time_per_epoch = elapsed_total / epoch
        remaining_epochs = num_epochs - epoch
        est_remaining_time = avg_time_per_epoch * remaining_epochs
        
        print(f"Epoch Time: {epoch_total_time:.1f}s, Estimated remaining: {est_remaining_time:.1f}s "
              f"({est_remaining_time/60:.1f} minutes)")
        
        # Update learning rate
        old_lr = optimizer.param_groups[0]['lr']
        scheduler.step(test_loss)
        new_lr = optimizer.param_groups[0]['lr']
        if old_lr != new_lr:
            print(f"Learning rate changed: {old_lr:.6f} -> {new_lr:.6f}")
        
        # Track metrics
        train_losses.append(train_loss)
        test_losses.append(test_loss)
        wers.append(wer)
        
        # Save model
        checkpoint_path = f'checkpoints/model_epoch_{epoch}.pt'
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'train_loss': train_loss,
            'test_loss': test_loss,
            'wer': wer,
            'char_map': char_map,
            'index_map': index_map
        }, checkpoint_path)
        print(f"Checkpoint saved: {checkpoint_path}")
        
        # Save best model
        if wer < best_wer:
            best_wer = wer
            best_checkpoint_path = 'checkpoints/model_best.pt'
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'train_loss': train_loss,
                'test_loss': test_loss,
                'wer': wer,
                'char_map': char_map,
                'index_map': index_map
            }, best_checkpoint_path)
            patience_counter = 0
            print(f"New best model saved with WER: {wer:.4f}")
        else:
            patience_counter += 1
        
        # Update overall progress
        overall_progress.update(1)
        '''
        # Early stopping
        if patience_counter >= config.early_stopping_patience:
            print(f"Early stopping after {epoch} epochs")
            break
            '''
    
    # Close progress bar
    overall_progress.close()
    
    # Print final summary with timing information
    total_time = time.time() - start_time
    print(f"\n{'='*50}")
    print(f"Training complete! Best WER: {best_wer:.4f}")
    print(f"Total training time: {total_time:.1f}s ({total_time/60:.1f} minutes)")
    print(f"{'='*50}")
    
    # Plot training history
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(test_losses, label='Test Loss')
    plt.legend()
    plt.title('Loss vs. Epoch')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    
    plt.subplot(1, 2, 2)
    plt.plot(wers, label='WER')
    plt.legend()
    plt.title('Word Error Rate vs. Epoch')
    plt.xlabel('Epoch')
    plt.ylabel('WER')
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()
    
except Exception as e:
    print(f"Error in training: {e}")
    raise e
```
